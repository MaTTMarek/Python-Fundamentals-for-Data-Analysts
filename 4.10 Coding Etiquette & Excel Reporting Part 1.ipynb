## The Instacart officers are interested in comparing customer behavior in different geographic areas.

## Create a regional segmentation of the data.

## You’ll need to create a “Region” column based on the “State” column from your customers data set.

### Creating Regions

Use the region information in this Wikipedia article to create your column (you only need to create regions, not divisions).

```python
# check state column
ords_custs_merged_2['State'].value_counts(dropna = False)
```

```python
# Creating a list of regions
region = []
for value in ords_custs_merged_2 ['State']:
 
    if (value in ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'Pennsylvania', 'New Jersey']):
        region.append('Northeast')
    elif (value in ['Wisconsin', 'Michigan', 'Illinois', 'Ohio', 'North Dakota', 'South Dakota', 'Nebraska', 'Minnesota', 'Iowa', 'Missouri']):
        region.append('Midwest')
    elif (value in ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana']):
        region.append('South')
    else:
        region.append('West')
```

```python
# adding new region column to data frame
ords_custs_merged_2 ['region'] = region
```

# Question 5.
## Create a profiling variable based on age, income, certain goods in the “department_id” column, and number of dependents.
### You might also use the “orders_day_of_the_week” and “order_hour_of_day” columns if you can think of a way they would impact customer profiles.
### (Hint: As an example, try thinking of what characteristics would lead you to the profile “Single adult” or “Young parent.”)

```python
# Checking age of customers

ords_custs_merged_2['Age'].value_counts().sort_index()
```

```python
# redistibuting age of customers
Age_distribution = []

for value in ords_custs_merged_2['Age']:
    if (value >=18) and (value <=25):
        Age_distribution.append('Age_18-25')
    elif (value >=26) and (value <=41):
        Age_distribution.append('Age_26-41')
    elif (value >=42) and (value <=57):
        Age_distribution.append('Age_42-57')
    elif (value >=58):
        Age_distribution.append('Age_58>')
```

```python
# adding new column
ords_custs_merged_2['Age_distribution'] = Age_distribution
```

```python
# checking
ords_custs_merged_2['Age_distribution'].value_counts(dropna = False)
```

```python
# grouping income
income_group = []
for value in ords_custs_merged_2 ['income']:
    if (value < 32000):
        income_group.append(' <32K')
    elif (value >= 32000) and (value <= 53000):
        income_group.append('32K-53K')
    elif (value > 53000) and (value <= 107000):
        income_group.append('53K-107K')
    elif (value > 107000) and (value <=374000):
        income_group.append('107K-374K')
    else:
        income_group.append('>374K')
    
```

```python
# adding new column with income
ords_custs_merged_2['income_group'] = income_group
```
