### Combining two data sets

```python
ords_prods_customers_merged = ords_prods_merged.merge(df_customers, on = 'user_id')
```

```python
# Creating day_of_week column with loc function 
ords_prods_customers_merged.loc[ords_prods_customers_merged['order_dow'] == 0, 'day_of_week'] = 'Saturday'
ords_prods_customers_merged.loc[ords_prods_customers_merged['order_dow'] == 1, 'day_of_week'] = 'Sunday'
ords_prods_customers_merged.loc[ords_prods_customers_merged['order_dow'] == 2, 'day_of_week'] = 'Monday'
ords_prods_customers_merged.loc[ords_prods_customers_merged['order_dow'] == 3, 'day_of_week'] = 'Tuesday'
ords_prods_customers_merged.loc[ords_prods_customers_merged['order_dow'] == 4, 'day_of_week'] = 'Wednesday'
ords_prods_customers_merged.loc[ords_prods_customers_merged['order_dow'] == 5, 'day_of_week'] = 'Thursday'
ords_prods_customers_merged.loc[ords_prods_customers_merged['order_dow'] == 6, 'day_of_week'] = 'Friday'
```
